```python
import pandas as pd

# Load data
file_path = "/Users/emilymoore/Downloads/DS 4002 Project 3/Breast-Lesions-USG-clinical-data.xlsx"
df = pd.read_excel(file_path)

# ---------------------------
# 1. Standardize column names
# ---------------------------
df.columns = df.columns.str.strip()

# ---------------------------
# 2. Keep only desired columns
# ---------------------------
desired_cols = [
    "Image_filename",
    "Mask_tumor_filename",
    "Age",
    "Tissue_composition",
    "Signs",
    "Symptoms",
    "BIRADS",
    "Shape",
    "Margin",
    "Echogenicity",
    "Posterior_features",
    "Halo",
    "Calcifications",
    "Skin_thickening",
    "Classification"
]

# Keep only columns that actually exist
df = df[[col for col in desired_cols if col in df.columns]]

# ---------------------------
# 3. Basic cleaning
# ---------------------------

# Remove duplicate rows
df = df.drop_duplicates()

# Replace empty strings with NaN
df = df.replace(r'^\s*$', pd.NA, regex=True)

# Strip whitespace from string columns
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip()

# ---------------------------
# 4. Fix data types
# ---------------------------

# Convert Age to numeric
if "Age" in df.columns:
    df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

# Optional: convert BIRADS to numeric if needed
if "BIRADS" in df.columns:
    df["BIRADS"] = pd.to_numeric(df["BIRADS"], errors="coerce")

# ---------------------------
# 5. Optional: standardize categorical values
# ---------------------------
# Example: make all categorical text lowercase
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.lower()

# ---------------------------
# 6. Save cleaned dataset
# ---------------------------
df.to_csv("/Users/emilymoore/Downloads/DS 4002 Project 3/cleaned_breast_lesions.csv", index=False)

# ---------------------------
# 7. Quick check
# ---------------------------
print(df.info())
print(df.head())
```
